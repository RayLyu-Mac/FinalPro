```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pandas import DataFrame
from sklearn.decomposition import PCA
from numpy import expand_dims
from sklearn.model_selection import train_test_split
```

```python
Temp=['300k','600k','800k']
step=40
for j in range(7):
    lower=600+(j-1)*step
    upper=600+j*step
    fig= plt.figure(figsize=(7,9))
    for i in Temp:
        df=pd.read_table('cube5-{}-30-2s-2V-b-Normalized.dat'.format(i))
        plt.plot(df['x'][lower:upper],df['y'][lower:upper],label=i)
        plt.legend()
```

```python
def index(df300,df600,df800):
    num=3
    Li=[df300,df600,df800]
    for k in range(3):
        for i in range(num):
            Li.insert(i+1,(Li[i]+Li[i+1])/2)
    for k in range(3):
        for i in range(num):
            Li.insert(-i-1,(Li[-i-1]+Li[-i-2])/2)
    #print(Li)
    Tem=[300,600,800]
    for k in range(3):
        for j in range(num):
            Tem.insert(j+1,(Tem[j]+Tem[j+1])/2)
    for k in range(3):
        for j in range(num):
            Tem.insert(-j-1,(Tem[-j-1]+Tem[-j-2])/2)
    return Li,Tem
```

```python
df300=[item for item in (pd.read_table('cube5-300K-30-2s-2V-b-Normalized.dat')['y'].values)]
df600=[item for item in (pd.read_table('cube5-600K-30-2s-2V-b-Normalized.dat')['y'].values)]
df800=[item for item in (pd.read_table('cube5-800K-30-2s-2V-b-Normalized.dat')['y'].values)]
del df300[270:1130]
del df600[270:1130]
del df800[270:1130]
df300=np.array(df300)
df600=np.array(df600)
df800=np.array(df800)
Li,Tem=index(df300,df600,df800)
Fin={}
for jj in range(len(Tem)):
    Fin['{}K'.format(Tem[jj])]=Li[jj]
df=DataFrame(Fin)
df.set_index('300K',inplace=True)
df.to_csv('Nox_PeakRemoved.csv')
```
